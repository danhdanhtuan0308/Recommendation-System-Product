```python
reader = Reader()
data = Dataset.load_from_df(data_userbase[['user_id', 'product_id', 'rating']], reader)
```

```python
algorithm = SVD()
results = cross_validate(algorithm, data, measures=['RMSE', 'MAE'],cv=5, verbose=True)
```

```python
results
```

```python
trainset = data.build_full_trainset()
algorithm.fit(trainset)
```

```python
# userId = 32 has seen some product:
userId = 32
df_select = data_userbase[(data_userbase['user_id'] == userId) & (data_userbase['rating'] >=4)]
df_select = df_select.set_index('product_id')
#df_select = df_select.join(df_title)['Name']
df_select.head(df_select.shape[0])
```

```python
df_score = data_userbase[["product_id"]]
```

```python
df_score
```

```python
# The following are the top 5 products to be recommended to the user with userId
# To recommend products  to the given user,
# we can sort the list of movies in decreasing order of predicted ratings
# and take the top N pproduct as recommendations:
df_score['EstimateScore'] = df_score['product_id'].apply(lambda x: algorithm.predict(userId, x).est) # est: get EstimateScore
df_score = df_score.sort_values(by=['EstimateScore'], ascending=False)
```

```python
df_score = df_score.drop_duplicates()
```
